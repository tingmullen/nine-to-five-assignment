## Aggregate Your Friends and Code some SQL

Put together the following:

1. How many employees were reported in 2016 in all industries? Round to the nearest whole number.

```python
SELECT TOP 1 ROUND(value, 0) AS total_employees_2016
FROM dbo.annual_2016 d
JOIN dbo.series s ON d.series_id = s.series_id
WHERE s.industry_code = '00000000' 
AND s.data_type_code = '01';
```

2. How many women employees were reported in 2016 in all industries? Round to the nearest whole number. 

```python
SELECT ROUND(value, 0) AS women_employees_2016
FROM dbo.annual_2016 a
JOIN dbo.series s ON a.series_id = s.series_id
WHERE s.data_type_code = '10' 
AND s.industry_code = '00000000';
```

3. How many production/nonsupervisory employees were reported in 2016? Round to the nearest whole number. 

```python
SELECT ROUND(SUM(value), 0) AS count_production_nonsupervisory_employee
FROM dbo.annual_2016 AS data
JOIN dbo.series AS series ON data.series_id = series.series_id
WHERE series.data_type_code = '06';
```

4. In January 2017, what is the average weekly hours worked by production and nonsupervisory employees across all industries?

```python
SELECT ROUND(AVG(value), 2) AS avg_weekly_payroll
FROM dbo.january_2017 AS data
JOIN dbo.series AS series ON data.series_id = series.series_id
WHERE series.data_type_code = '07';
```

5. What is the total weekly payroll for production and nonsupervisory employees across all industries in January 2017? Round to the nearest penny.

```python
SELECT ROUND(SUM(value), 2) AS total_weekly_payroll
FROM dbo.january_2017 AS data
JOIN dbo.series AS series ON data.series_id = series.series_id
WHERE series.data_type_code = '82';
```

6. In January 2017, for which industry was the average weekly hours worked by production and nonsupervisory employees the highest? Which industry was the lowest?

```python
-- Highest
SELECT TOP 1 i.industry_name, d.value AS max_hours
FROM dbo.january_2017 d
JOIN dbo.series s ON d.series_id = s.series_id
JOIN dbo.industry i ON s.industry_code = i.industry_code
WHERE s.data_type_code = '07'
ORDER BY d.value DESC;

-- Lowest
SELECT TOP 1 i.industry_name, d.value AS min_hours
FROM dbo.january_2017 d
JOIN dbo.series s ON d.series_id = s.series_id
JOIN dbo.industry i ON s.industry_code = i.industry_code
WHERE s.data_type_code = '07'
ORDER BY d.value ASC;
```

## Subqueries, Unions, Derived Tables, Oh My!

1. Write a query that returns the `series_id`, `industry_code`, `industry_name`, and `value` from the `january_2017` table but only if that value is greater than the average value for `annual_2016` of `data_type_code` 82.

```python
SELECT TOP 50 
    d.series_id, 
    s.industry_code, 
    i.industry_name, 
    d.value
FROM dbo.january_2017 d
JOIN dbo.series s ON d.series_id = s.series_id
JOIN dbo.industry i ON s.industry_code = i.industry_code
WHERE d.value > (

SELECT AVG(a.value)
FROM dbo.annual_2016 a
JOIN dbo.series s2 ON a.series_id = s2.series_id
WHERE s2.data_type_code = '82'
)
ORDER BY d.series_id;    
```

**Optional Bonus Question:** Write the above query as a common table expression!

```python
-- Optional CTE below
WITH average_payroll_2016 AS ( 
    SELECT AVG(a.value) AS avg_val
    FROM dbo.annual_2016 a
    JOIN dbo.series s2 ON a.series_id = s2.series_id
    WHERE s2.data_type_code = '82'
)
SELECT TOP 50 
    d.series_id, 
    s.industry_code, 
    i.industry_name, 
    d.value
FROM dbo.january_2017 d
JOIN dbo.series s ON d.series_id = s.series_id
JOIN dbo.industry i ON s.industry_code = i.industry_code
CROSS JOIN average_payroll_2016 
WHERE d.value > average_payroll_2016.avg_val
ORDER BY d.series_id;
```

2. Create a `Union` table comparing average weekly earnings of production and nonsupervisory employees between `annual_2016` and `january_2017` using the data type 30.  Round to the nearest penny.  You should have a column for the average earnings and a column for the year, and the period.

```python
SELECT 
    ROUND(AVG(d16.value), 2) AS avg_weekly_earnings, -- 别名改为全小写
    '2016' AS year, 
    'annual' AS period
FROM dbo.annual_2016 d16
JOIN dbo.series s ON d16.series_id = s.series_id
WHERE s.data_type_code = '30'

UNION

-- 2017 Jan data
SELECT 
    ROUND(AVG(d17.value), 2) AS avg_weekly_earnings, 
    '2017' AS year, 
    'january' AS period
FROM dbo.january_2017 d17
JOIN dbo.series s ON d17.series_id = s.series_id
WHERE s.data_type_code = '30';
```
